# PART ONE-

Here we're trying to understand the BLLs of children across USA and the deaths over the world due to lead exposure. The data sources that I have sourced my data from give an idea of the blood lead levels in children across various countries in the world during the year 2019 and in the different states of United States of America from 1996 to 2016.

<b> Variables of interest:
    
    

•	bll_usa for the mean blood lead levels in USA
    
•	bll_world for the blood lead levels of different countries in the world
    
•	bll_usa_state for the blood lead levels of different states in USA
    
•	bll_usa_state_meanleadlevels for the mean blood lead levels of different states in USA

    

```python
import pandas as pd
import numpy as np

#reading the different files to extract the data
bll_usa = pd.read_csv('lead-blood-usa-children.csv')
bll_world= pd.read_csv('lead-concentrations-blood-children-world.csv')
bll_usa_state= pd.read_csv('childhood-lead.csv')
```

```python
bll_world = bll_world[['Entity','Year', 'Blood lead level in children']]
bll_world
```

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# set plot style: grey grid in the background:
sns.set(style="darkgrid")

# Set the figure size(we choose 33 because of the number of countries in our data frame)
plt.figure(figsize=(33, 10))

# grouped barplot
#ax.set(ylim=(0, 100))
ax = sns.barplot(x = "Entity", y = "Blood lead level in children", data =bll_world )

ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
```

<b> Interpretation: 

From the above graph, our Palestine and Pakistan seem to have the highest levels of lead levels in children with Ethopia and Tanzania having the least levels.

```python
#we group the data in the states files by Year and State and calculate the cummulative counts of the percentage of children to get an idea
#of the recurative nature of higher incidences in each state
bll_summary = bll_usa_state[['Year', 'State', '% of Children Tested with Confirmed BLLs ≥10 µg/dL']]
bll_summary = bll_summary.groupby(['Year', 'State']).agg(Count = ('% of Children Tested with Confirmed BLLs ≥10 µg/dL', 'count')).reset_index()
bll_summary
```

```python
#We calculate the aggregate of blood lead levels in the different States while eliminating the NaN values
bll_usa_state.mean(axis=1, skipna= True)
```

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# set plot style: grey grid in the background:
sns.set(style="darkgrid")

# Set the figure size
plt.figure(figsize=(33, 10))

# grouped barplot
#ax.set(ylim=(0, 100))
ax = sns.barplot(x = "State", y = "% of Children Tested with Confirmed BLLs ≥10 µg/dL", data =bll_usa_state )

ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
```

```python
#now we calculate the average of lead levels across the rows for each state
bll_usa_state_meanleadlevels= bll_usa_state.loc[:,"5-9 µg/dL" :"≥70 µg/dL"].mean(axis=1)
```

```python
bll_usa_state_meanleadlevels
```

```python
#Now we add a column depicting these mean lead levels
bll_usa_state['bll_usa_state_meanleadlevels'] = bll_usa_state_meanleadlevels
```

```python
bll_usa_state
```

```python
#creating a revised data frame with the year as 2015, the different states and the mean lead levels
bll_usa_state = bll_usa_state[['Year','State', 'bll_usa_state_meanleadlevels']]
bll_usa_state
```

```python
#Now we plot this data frame in to a bar plot
import seaborn as sns
import matplotlib.pyplot as plt

# set plot style: grey grid in the background:
sns.set(style="darkgrid")

# Set the figure size
plt.figure(figsize=(33, 10))

# grouped barplot
#ax.set(ylim=(0, 100))
ax = sns.barplot(x = "State", y = "bll_usa_state_meanleadlevels", data =bll_usa_state )

ax.set_xticklabels(ax.get_xticklabels(), rotation = 90)
```

<b> Interpretation:
    
From the above data we can deduce that US totals which seems to be an aggregate of all the values is at the definitive highest. Statewise,Illinois and Ohio seem to be at a higher range compared to the rest of the states with the District of Columbia, Iowa, Kansas, Maine, Florida, Nevada, Houston only, Virginia and Washington being on the lower end of the spectrum.
